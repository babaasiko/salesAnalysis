```python
#5. Write a Pandas program to remove the duplicates from 'WHO region' column of World alcohol consumption dataset
# to view the duplicates
duplicates = acohol_data['WHO region'].duplicated().sum()
duplicates
```

```python
#to drop the duplicates
#syntax; df i.e data frame also known as variable name. df.duplicates(keep= 'first' or 'last'; this means python should not delete the first or the last)
acohol_data.drop_duplicates(subset = 'WHO region', keep='first')
```

```python
#6. Write a Pandas program to find out the alcohol consumption of a given year from the world alcohol consumption dataset
acohol_data.groupby('Year').sum()
```

```python
#7. Write a Pandas program to find out the alcohol consumption details in the year '1987' or '1989' from the world alcohol consumption dataset
acohol = acohol_data[acohol_data['Year'] == 1987]
acohol
```

```python
acohol = acohol_data[acohol_data['Year'] == 1989]
acohol
```

```python
#to view the consumption details for 1987 or 1989 using the | operator instead of 'or'
acohol = acohol_data[(acohol_data['Year'] == 1987) | (acohol_data['Year'] == 1989)]
acohol
```

```python
#8. Write a Pandas program to find out the alcohol consumption details by the 'Americas' in the year '1985' from the world alcohol consumption dataset
acohol = acohol_data[(acohol_data['WHO region'] == 'Americas') & (acohol_data['Year'] == 1985)]
acohol
```

```python
#9. Write a Pandas program to find out the alcohol consumption details in the year '1986' where WHO region is 'Western Pacific' and country is 
#'VietNam' from the world alcohol consumption dataset
acohol_data[(acohol_data['Year'] == 1986) & (acohol_data['WHO region']=='Western Pacific') & (acohol_data['Country']=='Viet Nam')]
```

```python
#10. Write a Pandas program to find out the alcohol consumption details in the year '1986' or '1989' where WHO region is 'Americas' from the world
#alcohol consumption dataset.
acohol = acohol_data[((acohol_data['Year'] == 1986) | (acohol_data['Year'] == 1989)) & ((acohol_data['WHO region']== 'Americas'))] 
acohol
```

```python
#11. Write a Pandas program to find out the alcohol consumption details in the year '1986' or '1989' where WHO region is 'Americas' or 'Europe' from the
#world alcohol consumption dataset
acohol = acohol_data[((acohol_data['Year'] == 1986) | (acohol_data['Year'] == 1989)) & ((acohol_data['WHO region']== 'Americas')|(acohol_data['WHO region']=='Europe'))] 
acohol
```

```python
#12. Write a Pandas program to find out the 'WHO region, 'Country', 'Beverage Types' in the year '1986' or '1989' where WHO region is 'Americas' or 'Europe'
#from the world alcohol consumption dataset.
acohol = acohol_data[((acohol_data[['WHO region', 'Country', 'Beverage Types']].isin == 1986) | (acohol_data[['WHO region', 'Country', 'Beverage Types']]== 1989)) & ((acohol_data['WHO region']== 'Americas')|(acohol_data['WHO region']=='Europe'))] 
acohol
```
